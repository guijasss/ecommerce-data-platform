```python
from pyspark.sql.functions import col, upper, to_timestamp
from pipelines.bronze.stream_utils import read_stream, write_stream

bronze_clicks_path = "/Volumes/main/bronze/ecommerce/clicks"
silver_clicks_path = "/Volumes/main/silver/ecommerce/clicks"

schema_location = "/Volumes/main/silver/ecommerce/_schemas/clicks"
checkpoint_location = "/Volumes/main/silver/ecommerce/_checkpoints/clicks"

stream_df = read_stream(
    source_path=bronze_clicks_path,
    schema_location=schema_location,
)

transformed_df = (
    stream_df
    .withColumn("occurred_at_ts", to_timestamp(col("occurred_at")))
    .withColumn("event_type_upper", upper(col("event_type")))
)

query = write_stream(
    df=transformed_df,
    output_path=silver_clicks_path,
    checkpoint_location=checkpoint_location,
    options={
        "partition_by": "event_date"
    }
)

query.awaitTermination()
```
